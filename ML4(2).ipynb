# Task
Construct a binomial distribution with different success rates to find out an estimated success rate for a product that received 97 positive reviews out of 100. Use histograms to assert your answer.

## Define parameters

### Subtask:
Define the number of trials (total reviews) and the number of successes (positive reviews).


**Reasoning**:
Define the variables for the number of trials and successes as instructed.



```
n_trials = 100
n_successes = 97
```

## Generate binomial distributions

### Subtask:
Generate binomial distributions with different success rates (probabilities of success) and the defined number of trials.


**Reasoning**:
Generate binomial distributions for different success rates using the defined number of trials.



```
from scipy.stats import binom

success_rates = [0.9, 0.95, 0.97, 0.99]
distributions = {}

for p in success_rates:
    distributions[p] = binom.rvs(n_trials, p, size=10000)
```

## Visualize distributions

### Subtask:
Plot histograms for each generated binomial distribution to visualize the probability mass function.


**Reasoning**:
Import the necessary library and iterate through the distributions to plot histograms as instructed.



```
import matplotlib.pyplot as plt

for p, dist in distributions.items():
    plt.figure()
    plt.hist(dist)
    plt.title(f'Binomial Distribution (p={p})')
    plt.xlabel('Number of Successes')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Highlight observed success

### Subtask:
Mark the observed number of successes (97) on each histogram to see which success rates are more likely.


**Reasoning**:
Iterate through the distributions, plot each histogram, and add a vertical line at the observed number of successes.



```
for p, dist in distributions.items():
    plt.figure()
    plt.hist(dist)
    plt.axvline(n_successes, color='red', linestyle='dashed', linewidth=2)
    plt.title(f'Binomial Distribution (p={p})')
    plt.xlabel('Number of Successes')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Estimate success rate

### Subtask:
Based on the histograms, estimate the success rate that best fits the observed data.

*   Marking the observed number of successes (97) on each histogram revealed that success rates of 0.95 and 0.99 had the peak of their distributions closer to 97, suggesting these rates are more likely to have produced the observed outcome compared to a success rate of 0.90.
