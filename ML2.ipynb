```python
import pandas as pd
import numpy
```

```python
df = pd.read_csv('faa_ai_prelim.csv')

df.head()
```

```python
df.info()
```

## Convert date and time columns

### Subtask:
Combine the `EVENT_LCL_DATE` and `EVENT_LCL_TIME` columns into a single datetime object and create a new column for this.


**Reasoning**:
Concatenate the date and time columns, convert the result to datetime objects, create a new column for these objects, and display the first few rows.



```python
df['EVENT_DATETIME_STR'] = df['EVENT_LCL_DATE'].fillna('') + ' ' + df['EVENT_LCL_TIME'].fillna('')
df['EVENT_DATETIME'] = pd.to_datetime(df['EVENT_DATETIME_STR'], format='%d-%b-%y %H:%M:%SZ', errors='coerce')
df.drop(columns=['EVENT_DATETIME_STR'], inplace=True)
df.head()
```

## Data selection

### Subtask:
Create a new dataframe with only the specified columns: `Aircraft make name`, `State name`, `Aircraft model name`, `Text information`, `Flight phase`, `Event description type`, and `Fatal flag`.


**Reasoning**:
I will create a new DataFrame `df_selected` containing only the specified columns, as requested by the user, and then display the first 5 rows of this new DataFrame.



```python
required_columns = ['ACFT_MAKE_NAME', 'LOC_STATE_NAME', 'ACFT_MODEL_NAME', 'RMK_TEXT', 'FLT_PHASE', 'EVENT_TYPE_DESC', 'FATAL_FLAG']
df_selected = df[required_columns]
df_selected.head()
```

## Handle missing values in fatal flag

### Subtask:
Fill missing values in the `Fatal Flag` column with a suitable value (e.g., 'No' or 'Unknown').


**Reasoning**:
Fill the missing values in the `FATAL_FLAG` column with 'No' and verify the changes.



```python
# Recreate df_selected with all necessary columns
required_columns = ['ACFT_MAKE_NAME', 'LOC_STATE_NAME', 'ACFT_MODEL_NAME', 'RMK_TEXT', 'FLT_PHASE', 'EVENT_TYPE_DESC', 'FATAL_FLAG', 'ACFT_DMG_DESC']
df_selected = df[required_columns].copy()

# Impute missing values
df_selected['FATAL_FLAG'] = df_selected['FATAL_FLAG'].fillna('No')
flt_phase_mode = df_selected['FLT_PHASE'].mode()[0]
df_selected['FLT_PHASE'] = df_selected['FLT_PHASE'].fillna(flt_phase_mode)
acft_dmg_mode = df_selected['ACFT_DMG_DESC'].mode()[0]
df_selected['ACFT_DMG_DESC'] = df_selected['ACFT_DMG_DESC'].fillna(acft_dmg_mode)

print("Missing values after imputation:")
print(df_selected[['FATAL_FLAG', 'FLT_PHASE', 'ACFT_DMG_DESC']].isnull().sum())

# Drop rows with missing aircraft make name
df_selected.dropna(subset=['ACFT_MAKE_NAME'], inplace=True)
print("\nNumber of rows after dropping missing 'ACFT_MAKE_NAME':", df_selected.shape[0])

# One-hot encode ACFT_DMG_DESC
acft_dmg_dummies = pd.get_dummies(df_selected['ACFT_DMG_DESC'], prefix='ACFT_DMG', drop_first=True)
df_selected = pd.concat([df_selected, acft_dmg_dummies], axis=1)
df_selected.drop('ACFT_DMG_DESC', axis=1, inplace=True)

print("\ndf_selected columns after one-hot encoding and dropping original:")
print(df_selected.columns.tolist())

# Group by aircraft name and count
acft_make_groups = df_selected.groupby('ACFT_MAKE_NAME')
acft_make_counts = acft_make_groups.size()

print("\nAircraft make counts:")
display(acft_make_counts)

# Display observations with fatal flag "yes"
fatal_incidents = df_selected[df_selected['FATAL_FLAG'] == 'Yes']
print("\nFatal incidents:")
display(fatal_incidents)
```

## Summary:

### Data Analysis Key Findings

*   The dataset initially contained 83 observations.
*   After dropping rows with missing aircraft make names, 78 observations remained.
*   The "FATAL\_FLAG" column had missing values which were filled with 'No'.
*   The "FLT\_PHASE" and "ACFT\_DMG\_DESC" columns also had missing values, which were filled with the mode of each respective column.
*   All columns in the selected dataframe had more than 75 non-null values, leading to all of them being dropped in one step. This required recreating the selected dataframe to proceed with subsequent analysis.
*   The most frequent aircraft makes in the dataset are 'CESSNA' (23 occurrences), 'PIPER' (10 occurrences), and 'BEECH' (9 occurrences).
*   There were 8 observations identified as fatal incidents in the filtered data.
*   One-hot encoding was applied to the "ACFT\_DMG\_DESC" column, creating new binary columns for different damage descriptions.
